```python
from sklearn.feature_selection import mutual_info_classif as MIC
mi_score = MIC(x,y)
print(mi_score)
```

```python
def make_mi_scores(x, y):
    mi_scores = MIC(x, y)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=x.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(x, y)
mi_scores
```

```python
discrete_features = x.dtypes == int
def make_mi_scores(x, y, discrete_features):
    mi_scores = MIC(x, y, discrete_features=discrete_features)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=x.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores

mi_scores = make_mi_scores(x, y, discrete_features)
mi_scores
```

```python
from sklearn.metrics import confusion_matrix, classification_report,accuracy_score
```

```python
def gen_metrics(model,x_train,x_test,y_train,y_test):
    model.fit(x_train,y_train)
    train_score = model.score(x_train,y_train)
    test_score = model.score(x_test,y_test)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test,y_pred)
    print('Prediction\n',y_pred)
    print('Training score',train_score)
    print('Testing score',test_score)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test,y_pred)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test,y_pred))
```

```python
def gen_met_pca(model,x_train1,x_test1,y_train1,y_test1):
    model.fit(x_train1,y_train1)
    train_score = model.score(x_train1,y_train1)
    test_score = model.score(x_test1,y_test1)
    y_pred1 = model.predict(x_test1)
    acc = accuracy_score(y_test1,y_pred1)
    print('Prediction\n',y_pred1)
    print('training score',train_score)
    print('testing score',test_score)
    print('Accuracy_Score',acc)
    cm = confusion_matrix(y_test1,y_pred1)
    print('Confusion Matrix\n',cm)
    print('Classification Report\n',classification_report(y_test1,y_pred1))
```
